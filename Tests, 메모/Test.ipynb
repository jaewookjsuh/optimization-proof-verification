```python
class CustomNumber:
    def __init__(self, value):
        self.value = value

    def __eq__(self, other):
        return self.value == other.value

    def __setattr__(self, name, value):
        print(f"Setting {name} to {value}")
        # 이렇게 하면 모든 속성에 대해 할당할 때마다 메시지가 출력됩니다.
        super().__setattr__(name, value)

# 사용자 정의 클래스 객체 생성
num1 = CustomNumber(10)
num2 = CustomNumber(20)

# __eq__ 메서드를 사용하여 동등성 비교
print(num1 == num2)  # 출력: False

# __setattr__ 메서드에 따른 동작
num1.value = 30
# 출력: Setting value to 30
```

# Min test

```python
def is_iterable(obj):
    try:
        iter(obj)
        return True
    except TypeError:
        return False
```

```python
class Min(Scalar):
    def __init__(self, min_set):
        self.base = self
        self.min_set = min_set
        if is_iterable(min_set):
            for element in min_set:
                Prop.add( le(self, element) )

    def __str__(self):
        s = 'Min('
        if is_iterable(self.min_set):
            for element in self.min_set:
                s += str(element) + ', '
            s = s[:-2]
        s += ')'
        return s

    def __repr__(self):
        s = '\\text{Min} \left('
        if is_iterable(self.min_set):
            for element in self.min_set:
                s += repr(element) + ', \,\,'
            s = s[:-6]
        s += '\\right)'
        return s
```

```python
a = Scalar('a')
b = Scalar('b')

min_set = set([a,b])

m = Min(min_set)
```

```python
verify(m<=a, set())
```
